```python
import matplotlib.pyplot as plt
```

```python
y_vector = [0.14, 0.562, 1.25, 2.203, 3.452, 4.968, 6.779, 9.123, 11.248, 13.764]
```

```python
y_list = [0.234, 1.062, 2.593, 4.906, 7.779, 11.544, 15.86, 21.395, 27.461, 34.178]
```

```python
plt.plot(x, y_vector)
plt.plot(x, y_list)
plt.xlabel('Number of elements (K)')
plt.ylabel('Time (s)')
plt.legend(("vector", "list"), loc = "upper left")
plt.show()
```
