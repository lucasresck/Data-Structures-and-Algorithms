```python
import matplotlib.pyplot as plt
```

```python
x = list(range(10, 101, 10))
```

```python
y_vector = [140, 562, 1235, 2234, 3546, 5034, 6857, 8936, 11312, 13952]
for i in range(len(y_vector)):
    y_vector[i] /= 1000
```

```python
y_list = [218, 1047, 2594, 4811, 7850, 11388, 16014, 21042, 26785, 33452]
for i in range(len(y_list)):
    y_list[i] /= 1000
```

```python
plt.plot(x, y_vector)
plt.plot(x, y_list)
plt.xlabel('Number of elements (K)')
plt.ylabel('Time (s)')
plt.legend(("vector", "list"), loc = "upper left")
plt.show()
```
